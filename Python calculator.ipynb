```python
# Addition
def add(x,y):
    return x+y

# Subtraction
def subtract(x,y):
    return x-y

# Multiplication
def multiply(x,y):
    return x*y

# Division
def divide(x,y):
    return x/y

# Power
def power(x, y):
    return x**y

# Root
def root(x):
    return x**0.5

print("Choose operation.")
print("1.Add")
print("2.Subtract")
print("3.Multiply")
print("4.Divide")
print("5.Power")
print("6.Root")

while True:
    # Take user Input
    choice = input("Enter choice(1/2/3/4/5/6): ")

    # Check if choice is above option
    if choice in ('1', '2', '3', '4','5'):
        try:
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue
        # Choices for result
        if choice == '1':
            print(num1,"+",num2,"=",add(num1,num2))
        elif choice == '2':
            print(num1,"-",num2,"=",subtract(num1,num2))
        elif choice == '3':
            print(num1,"*",num2,"=",multiply(num1,num2))
        elif choice == '4':
            print(num1,"/",num2,"=",divide(num1,num2))
        elif choice == '5':
            print(num1,"^",num2,"=",power(num1,num2))
            # Continuation of claculator
        cont = input("Do you want to calculate again? (Type \'n\' to terminate calculator)")
        if cont =='n':
            break
    
    # Root and potential exit
    elif choice in ('6',"exit"):
        try:
            num1 = float(input("Enter number: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue
        if choice == '6':
            print("âˆš",num1,"=",root(num1))
        # Continuation for calculator again
        cont = input("Do you want to calculate again? (Type \'n\' to terminate calculator)")
        if cont =='n':
            break
    else:
        print("Invalid Input")
```
